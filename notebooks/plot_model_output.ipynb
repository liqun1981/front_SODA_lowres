```python
# Import packagaes
import xmitgcm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set path to model output
path_data0 = '/work/Marion.Alberty/front_SODA_lowres/run_day1'
path_dataN = '/work/Marion.Alberty/front_SODA_lowres/run_day4_7'
# Open dataset with xmitgcm
ds0 = xmitgcm.open_mdsdataset(path_data0,iters=[0],geometry='cartesian')
dsN = xmitgcm.open_mdsdataset(path_dataN,iters=[604800],geometry='cartesian')
```

```python
# ds0
```

```python
# Plot initial sea surface height
ds0.Eta.plot()
plt.show()

# Plot final sea surface height
dsN.Eta.plot()
plt.show()
```

```python
# Plot Initial Surface Temperature
ds0.T.isel(Z = 0).plot()
plt.show()

# Plot Final Surface Temperature
dsN.T.isel(Z = 0).plot()
plt.show()
```

```python
# Plot Initial Salinity Section
ds0.S.isel(XC = 0).plot(levels=np.linspace(26,dsN.S.isel(XC = 0).max(),100))
plt.show()

# Plot Final Salinity Section
dsN.S.isel(XC = 0).plot(levels=np.linspace(26,dsN.S.isel(XC = 0).max(),100))
plt.show()
```

```python
# Plot Initial Temperature Section
ds0.T.isel(XC = 0).plot()
plt.show()

# Plot Final Salinity Section
dsN.T.isel(XC = 0).plot()
plt.show()
```

```python
# Plot Initial along front velocity
ds0.U.isel(XG = 10).plot()
plt.show()

# Plot Final along front velocity
dsN.U.isel(XG = 10).plot()
plt.show()
```

```python
# Plot Initial across-front velocity
ds0.V.isel(XC = 5).plot()
plt.show()

# Plot Final across-front velocity
dsN.V.isel(XC = 5).plot()
plt.show()
```

```python
# Plot Initial vertical velocity
ds0.W.isel(XC = 10).plot()
plt.show()
# Plot Final vertical velocity
dsN.W.isel(XC = 10).plot()
plt.show()
```
